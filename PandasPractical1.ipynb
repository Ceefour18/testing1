**PLOTTING USING PANDAS**

Pandas uses the **plot()** method to create diagrams.

But we need **Pyplot** submodule in **Matplotlib library** to visualize the diagram on the screen

```python
import matplotlib.pyplot as plt
df.plot(kind = 'scatter', x = 'total_rooms', y = 'total_bedrooms')

plt.show()
```

```python

df.plot(kind = 'hist', x = 'total_rooms', y = 'total_bedrooms')

plt.show()
```

```python
df.plot(kind = 'bar', x = 'longitude', y = 'latitude')
plt.show()
```

```python
df[:10][['longitude', 'latitude']].plot(kind = 'bar', x = 'longitude', y = 'latitude')
plt.show()
```

Here are some of the chart types you can plot in Pandas, along with their syntax:

**1. Line Chart**

df.plot(kind = 'line', x = 'column1', y = 'column2')

**2. Scatter Plot**

df.plot(kind = 'scatter', x = 'column1', y = 'column2')
plt.show()


**3. Bar Chart**

df.plot(kind = 'bar', x = 'column1', y = 'column2')
plt.show()

**4. Histogram**

df.plot(kind = 'hist', x = 'column1', y = 'column2')
plt.show()


**5. Box Plot**

df.plot(kind = 'box')
plt.show()

**6. Area Plot**
df.plot(kind = 'area')
plt.show()

**7. Pie Chart**
